```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
df_train, df_test = train_test_split(df, test_size = 0.2, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
num_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking','price','areaperbedroom']
scaler.fit(df_train[num_vars])
df_train[num_vars] = scaler.transform(df_train[num_vars])
df_test[num_vars] = scaler.transform(df_test[num_vars])
df_train.head()
```

```python
plt.figure(figsize=(16,10))
sns.heatmap(df_train.corr(),annot = True)
plt.show()
```

```python
# # dividing into x and y
y_train = df_train.pop('price')
X_train = df_train
y_test = df_test.pop('price')
X_test = df_test
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

**Using K-Fold cross validation to measure accuracy of our LinearRegression model** 

```python
from matplotlib.patches import Shadow
from sklearn.model_selection import ShuffleSplit, cross_val_score
cv = ShuffleSplit(n_splits=5, test_size =0.2, random_state=0)
temp_X = pd.DataFrame(np.vstack((X_train,X_test)))
temp_y = pd.DataFrame(np.hstack((y_train,y_test)))
cross_val_score(LinearRegression(), temp_X,temp_y,cv=cv)
```

**Find best Model using GridSearchCV**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(temp_X,temp_y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'normalize': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [0.1,0.5,0.01,0.001,0.0001],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(temp_X,temp_y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })
    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(temp_X,temp_y)
```
