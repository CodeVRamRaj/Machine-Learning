```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
datasets = pd.read_csv('X_train.csv')
```

```python
datasets.head()
```

```python
y_train = datasets[['SalePrice']]  #dependent feature
X_train = datasets.drop(['Id','SalePrice'],axis=1)
```

```python
# Apply Freature Selection
feature_sel_model = SelectFromModel(Lasso(alpha=0.005,random_state=0))
feature_sel_model.fit(X_train,y_train)
```

```python
#here we select less value of alpha... because if we select high value then less features that will be selected
```

```python
selected_feat = X_train.columns[(feature_sel_model.get_support())]
```

```python
X_train = X_train[selected_feat] 
X_train
```

```python
from sklearn.model_selection import train_test_split
X_train1,X_test1,y_train1,y_test1 = train_test_split(X_train,y_train,random_state=0)
```

```python
lassoReg = Lasso().fit(X_train1,y_train1)
print("Training set score: {:.2f}".format(lassoReg.score(X_train1,y_train1)))
print("Test set score: {:.2f}".format(lassoReg.score(X_test1,y_test1)))
print("Number of feature used: {:.2f}".format(np.sum(lassoReg.coef_ !=0)))
```
