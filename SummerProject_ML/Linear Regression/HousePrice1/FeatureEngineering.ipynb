## 2-> Feature Engineering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

datasets = pd.read_csv("train.csv")
datasets.head()
```

```python
#categorical data ...nan values
features_nan_cat=[feature for feature in datasets.columns if datasets[feature].isnull().sum()>1 and datasets[feature].dtypes=='O']
datasets[features_nan_cat]=datasets[features_nan_cat].fillna('Missing')
```

```python
datasets.head()
```

```python
features_nan_num=[feature for feature in datasets.columns if datasets[feature].isnull().sum()>1 and datasets[feature].dtypes!='O']
for feature in features_nan_num:
    median_value=datasets[feature].median()
    datasets[feature+'nan']=np.where(datasets[feature].isnull(),1,0)
    datasets[feature].fillna(median_value,inplace=True)
    
```

```python
datasets.head()
```

```python
datasets[features_nan_num].isnull().sum()    #to verify
    
```

```python
## Temporal Variables (Date Time Variables)
features_yr = [feature for feature in datasets.columns if 'Yr' in feature or 'Year' in feature]
features_yr
```

```python
features_yr = [feature for feature in features_yr if 'nan' not in feature]
features_yr.remove('YrSold') 
features_yr
```

```python
for feature in features_yr:
   datasets[feature]=datasets['YrSold']-datasets[feature]
datasets.head()
```

```python
datasets[features_yr]
```

```python
features_num = [feature for feature in datasets.columns if datasets[feature].dtypes !='O']
features_dis=[feature for feature in features_num if len(datasets[feature].unique())<25 and feature not in features_yr+['Id']]
features_con = [feature for feature in features_num if feature not in features_yr + ['Id']+ features_dis]
features_con
```

```python
datasets['1stFlrSF'].skew()
```

```python
#if categorical variables less than 1%...remove
features_cat=[feature for feature in datasets.columns if datasets[feature].dtype=='O']
for feature in features_cat:
    data=datasets.groupby(feature)['SalePrice'].count()/len(datasets)
    data_df=data[data>0.01].index
    datasets[feature]=np.where(datasets[feature].isin(data_df),datasets[feature],'Rare_var')
    
```

```python
for feature in features_cat:
    labels_ordered=datasets.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    datasets[feature]=datasets[feature].map(labels_ordered)
```

```python
datasets.head()
```

## Feature Scaling 

```python
feature_scale = [feature for feature in datasets.columns if feature not in ['Id','SalePrice']]
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler() 
scaler.fit(datasets[feature_scale])
```

```python
scaler.transform(datasets[feature_scale])
```

```python
train_set = pd.concat([datasets[['Id','SalePrice']].reset_index(drop=True),
           pd.DataFrame(scaler.transform(datasets[feature_scale]),columns=feature_scale)],axis=1)
```

```python
train_set
```

```python
train_set.to_csv('X_train.csv',index=False)
```
