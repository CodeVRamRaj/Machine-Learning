```python
from sklearn.datasets import *
from dtreeviz.trees import *
from IPython.display import Image, display_svg, SVG
```

**For Classification**


```python
clas = tree.DecisionTreeClassifier(max_depth=2)
iris = load_iris()
X_train = iris.data
y_train = iris.target
clas.fit(X_train, y_train)
```

```python
from sklearn.tree import plot_tree
plot_tree(clas)
```

```python
viz = dtreeviz(clas, X_train, y_train, feature_names = iris.feature_names, class_names =['setosa', 'versicolor', 'virginica'], scale=2) #orientation='LR'
viz
```

**Regression**

```python
regr = tree.DecisionTreeRegressor(max_depth=2)
boston = load_boston()
X_train = boston.data
y_train = boston.target
regr.fit(X_train, y_train)
viz = dtreeviz(regr, X_train,y_train, target_name ='price', feature_names = boston.feature_names)
viz
```

**Show prediction_path**

```python
clas = tree.DecisionTreeClassifier()
iris = load_iris()
X_train = iris.data
y_train = iris.target
clas.fit(X_train,y_train)
X = iris.data[np.random.randint(0, len(iris.data)),:]
viz = dtreeviz(clas, X_train, y_train, feature_names = iris.feature_names, class_names =['setosa', 'versicolor', 'virginica'],X=X, scale=1) #orientation='LR'
viz
```

```python
X
```

**Show Node Number**

```python
viz = dtreeviz(clas, X_train, y_train, feature_names = iris.feature_names, 
               class_names =['setosa', 'versicolor', 'virginica'], orientation='LR',show_node_labels=True) #orientation='LR'
viz
```

**Without any Graphs**

```python
viz = dtreeviz(clas, X_train, y_train, feature_names = iris.feature_names, 
               class_names =['setosa', 'versicolor', 'virginica'], orientation='LR',fancy=False) #orientation='LR'
viz
```

**Show just Prediction path**

```python
viz = dtreeviz(clas, X_train, y_train, feature_names = iris.feature_names, 
               class_names =['setosa', 'versicolor', 'virginica'], orientation='LR',X=X, show_just_path=True) 
viz
```

**Prediction Path in Plain English**
