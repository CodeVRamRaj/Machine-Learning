```python
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
```

```python
X,y = make_classification(n_samples=100, n_features=20, n_informative=5, n_redundant=0, n_repeated=15, n_classes=2,shuffle=True, random_state=None)
```

```python
from sklearn.model_selection import train_test_split 
import time
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=2, test_size=0.2)
obj = DecisionTreeClassifier()
s = time.time()
obj.fit(X_train, y_train)
print(time.time()-s)
y_predict = obj.predict(X_test)
```

```python
obj1 = DecisionTreeClassifier(criterion='entropy')
ss = time.time()
obj1.fit(X_train,y_train)
print(time.time()-ss)
y_pred = obj1.predict(X_test)
```
